# list

시퀀스 자료형이고, 여러개의 값을 제한없이 저장할 수 있는 자료형(데이터 구조)(순서를 유지)


## 생성/수정

```python
list1 = [1,2,3]
print(list1, type(list1))

list2 = list('Hello')
print(list2, type(list2))
```

```python
# list는 mutable(가변)한 자료형
nums = [1,2,3]
print(nums)

# 맨 뒤에 요소 추가
nums.append(4)  # inplace연산 -> 원본이 바뀜
print(nums)

# 특정자리에 요소 추가
nums.insert(1, 1.5)
print(nums)

# list의 요소는 모든 타입이다.
values = [1,2,3, "안녕", [10,20,30]]
print(values)
print(values[0])
print(values[3])
print(values[-1])
print(values[-1][1])
print(values[3][0])

print(values[1:3])
print(values[::-1])
```

# list 연산

```python
a = [1,2,3]
b = [4,5,6]
c = a + b # 연결연산
print(c)

d = a * 2
print(d)

# 특정요소에 대한 연산
a = [10, 20, 30]

print(a[1] * 10)
a[1] = a[1] * 10
print(a)

# slicing을 통한 값 설정
a = [10, 20, 30, 40, 50]
print(a[1:3])
a[1:3] = [7,8,9]    # 끼워넣기가 아닌 제거 후 추가
print(a)

a[3:] = []
print(a)
```

# list api

```python
# str.split(구분자)
data = "신사임당,여,강남구 역삼동"

data = data.split(',')
print(data)
print(f'{data[0]}({data[1]}): {data[2]} 거주중...')

# list.index()

nums = [10, 20, 30, 40, 10]
print(nums.index(10))
print(nums.index(10, 1))
print(nums.index(10, -1))

# print(nums.index(100))    # 100 is not in list

print(10 in nums)

if 10 in nums:
    print(nums.index(10))
```

```python
# list.count(value)
nums = [10,20,10,30,40,10,100]
print(nums.count(10))
print(nums.count(100))

# list.sort()
scores = [100,97,49,55,60]

scores.sort(reverse=True)
print(scores)

fruits = ['orange', 'apple', 'pear', 'banana', 'kiwi', 'apple']
fruits.sort(reverse=True)
print(fruits)

# 정렬기준을 변경: key 속성 함수 전달
fruits.sort(key=len)
print(fruits)
```

```python
# sort의 반환값은 None, 값없음을 의미
print(fruits.sort(), type(fruits.sort()))

fruits_ = sorted(fruits, reverse=True, key=len)
print(fruits_)
```

#  반복순회

list는 iterable객체로써, 요소를 반복순회 할 수 있다.

```python
foods = ['🥨', '🥓', '🍙', '🍜','🥞']

for food in foods:
    print(food)

# enumerate() 인덱스와 함께 요소를 반환
for k, food in enumerate(foods):
    print(k, food)

print(enumerate(foods))
```
