# set

요소의 중복을 허용하지 않는 자료형

```python
str1 = {3, 2, 1, 1, 4, 4, 2}
print(str1, type(str1))

str2 = {}
print(type(str2))
```

```python
list = [1,2,3,3,2,4,5] #list 중복허용
print(list)

st2 = set(list)
print(st2, type(st2))

tpl = ('a', 'b', 'b', 'a', 'b')
print(tpl)

st3 = set(tpl)
print(st3)

print(tuple(st3))
```

# 요소 추가/삭제

```python
nums = {1,2,3}
nums.add(4)
print(nums)

nums.remove(2)
print(nums)

# nums.remove(100)
nums.discard(100) # 존재하지 않는 값을 삭제하려해도 KeyError 발생 X
```
