# tuple

시퀀스 객체이면서 불변(immutable)의 객체 순서를 저장

```python
t1 = ()     # 빈 튜플
print(t1, type(t1))

t2 = (10,)
print(t2, type(t2))

t3 = 10, 20

print(t3, type(t3))
```

```python
tpl = 1,2,3
print(tpl)
print(tpl[0])
print(tpl[0:2])

print(tpl * 2)
print(tpl)

# tpl[0] = 100 'tuple' object does not support item assignment
# tpl[0:2] = (100,200)
```

```python
hong = ('홍길동', 33, [100, 89, 50])

# hong[0] = '신사임당'04_str.ipynb

hong[2][2] = 70
print(hong)
```

```python
# 값 치환
x = 33
y = 55

# temp = x
# x = y
# y = temp

x, y = y,x

print(f'x= {x}, y= {y}')
```

```python
foods = ['🥨', '🥓', '🍙', '🍜','🥞']

for food in foods:
    print(food)

# enumerate() 인덱스와 함께 요소를 반환
for k, food in enumerate(foods):
    print(k, food)
```
