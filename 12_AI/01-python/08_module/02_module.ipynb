# 모듈 module

- .py파일을 의미
- 코드 재사용성을 높이기 위해 모듈단위로 관련있는 것들을 묶어서 관리
- 모듈 하위의 변수, 함수, 클래스 등을 외부에서 임포트 후 사용 가능
- private 변수 선언시에는 외부에서 접근 불가

```python
import math

print(dir(math)) # 특정모듈의 사용가능한 변수들
print(dir())    # 현재모듈의 사용가능한 변수
```

```python
# 현재  모듈 이름 
print(__name__)

# import한 모듈 이름
print(math.__name__)
```
