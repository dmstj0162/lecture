# 파일 쓰기


```python
text = """안녕하세요 파이썬으로 만든 귀도 입니다.
I love you
123456789
🍔🍕🧇
"""

f = open('sample.txt', 'w', encoding='utf-8')
print(f, type(f))

f.write(text)
f.close() # 반드시 객체를 반환해야 메모리 누수가 생기지 않는다.
```

# 파일 읽기

```python
f = open('sample.txt', 'r', encoding='utf-8')

print(f.read())

f.close()
```

```python
# with블럭: 생성된 객체를 자동으로 반환
with open('sample.txt', 'w', encoding='utf-8') as f:
    f.write(text)

with open('sample.txt', 'r', encoding='utf-8') as f:
    print(f.read())
```

```python
# for문을 통해 한줄씩 읽기
with open('sample.txt', 'r', encoding='utf-8') as f:
    for line in f:
        print(line, end='')
```

```python
# readline(): 한 번에 한줄씩 읽어 반환
with open('sample.txt', 'r', encoding='utf-8') as f:
    # := 바다코끼리 연산자: 대입연산자 + 좌항값반환
    while line := f.readline():
        print(line, end='')
```

```python
# 이어쓰기
with open('sample.txt', 'a', encoding='utf-8') as f:
    f.write('Thank you\n')
```
